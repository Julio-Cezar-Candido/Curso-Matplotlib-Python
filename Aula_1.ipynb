## Fazendo as primeiras visualizações

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
df.head()
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
plt.plot(df['data'], df['temperatura'])
```

```python
plt.figure(figsize=(15,7))
plt.plot(df['data'], df['temperatura'])
```

```python
plt.figure(figsize=(15,7))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura no Tempo')
```

```python
fig = plt.figure(figsize=(15,7))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento')
```

```python
fig = plt.figure(figsize=(15,7))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento')
eixo.set_ylabel('Temperatura (K)')
eixo.set_xlabel('Data')
```

```python
fig = plt.figure(figsize=(15,7))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento', size=20)
eixo.set_ylabel('Temperatura (K)', size=16)
eixo.set_xlabel('Data', size=16)
```

```python
fig = plt.figure(figsize=(15,7))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento', size=20)
eixo.set_ylabel('Temperatura (K)', size=16)
eixo.set_xlabel('Data', size=16)
eixo.legend(['Temperatura'], loc='lower right', fontsize=14)
```

```python
fig = plt.figure(figsize=(15,7))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color='orange')

eixo.set_title('Temperatura no momento', size=20)
eixo.set_ylabel('Temperatura (K)', size=16)
eixo.set_xlabel('Data', size=16)
eixo.legend(['Temperatura'], loc='lower right', fontsize=14)
```
